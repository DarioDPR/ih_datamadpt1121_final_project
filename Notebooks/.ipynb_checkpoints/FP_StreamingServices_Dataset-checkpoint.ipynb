```python
#librerias
import pandas as pd
import requests
import bs4  
import re
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0'}

url = 'https://www.justwatch.com/es/proveedor/netflix/peliculas'
response = requests.get(url, headers=headers)
html = response.content
print(f'Status code is: {response.status_code}')
print(type(html))
```

```python
html[:20000]
```

```python
parsed_html = bs4.BeautifulSoup(html, "html.parser")
parsed_tags = set([tag.name for tag in parsed_html.find_all(True)])
parsed_tags
```

```python
element = parsed_html.find_all("a", {"title-list-grid__item--link"})
print(element[0])
links = [tag.attrs for tag in element]
links
```

```python
netflix = pd.DataFrame(links, columns=['href'])
netflix
```
