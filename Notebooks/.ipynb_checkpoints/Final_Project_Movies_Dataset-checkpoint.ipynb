```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_films_with_a_100%25_rating_on_Rotten_Tomatoes'
html = requests.get(url).content
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
table = soup.find_all('table',{'class':'wikitable'})[0]
table
```

```python
rows = table.find_all('tr')
rows = [row.text.strip().split('\n') for row in rows]
rows
```

```python
colnames = rows[0]
data = rows[1:]

rotten100 = pd.DataFrame(data, columns=colnames)
rotten100
```

```python
rotten100.sort_values(by=['No. of reviews'], inplace=True, ascending=False)
rotten100
```

```python
imdbrat = pd.read_csv('../data/ratings.tsv', sep='\t')
imdbrat.sort_values(by=['averageRating'], inplace=True, ascending=False)
imdbrat
```

```python
imdbtit = pd.read_csv('../data/title.tsv', sep='\t')
imdbtit
```

```python
imdbtit['titleType'].unique()
```

```python
imdbmov = imdbtit.loc[imdbtit['titleType'] == 'movie']
imdbmov
```

```python
imdb = pd.merge(imdbrat, imdbmov, on="tconst")
imdb.sort_values(by=['numVotes'], inplace=True, ascending=False)
imdb
```

```python
imdbtop = imdb.drop(imdb.index[np.where(imdb['averageRating'] < 7.0)])
imdbtop.sort_values(by=['averageRating'], inplace=True, ascending=False)

imdbtop
```

```python
imdbtop2 = imdbtop.drop(imdbtop.index[imdbtop['numVotes'] < 25000])
imdbtop2
```

```python
imdbtop3 = imdbtop.drop(imdbtop.index[imdbtop['numVotes'] < 200000])
imdbtop3
```
